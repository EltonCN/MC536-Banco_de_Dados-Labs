```python
import requests
import json

url = "https://opendata.arcgis.com/datasets/9f70a4da50f14906839a8a070bc0d665_1.geojson" #

response = requests.get(url)
data = response.json()
```

```python
data
```

```python
data["features"][0]["properties"]
```

```python
from matplotlib import pyplot as plt

hosp = {}

for feature in data["features"]:
    properties = feature["properties"]

    if properties["YEAR"] not in hosp:
        hosp[properties["YEAR"]] = 0

    if (properties["HOSP"] is not None and 
        properties["COUNTY"] == "Statewide" and 
        properties["GENDER"] == "Both genders" and
        properties["AGE"] == "All ages"):

        hosp[properties["YEAR"]] += properties["HOSP"]

year = list(hosp.keys())
hosp = list(hosp.values())

plt.suptitle("Hospitalização por asma")
plt.title("Estado do Colorado")
plt.plot(year, hosp)
plt.ylim(0, np.max(hosp))
plt.xlabel("Ano")
plt.ylabel("Hospitalização")
```
